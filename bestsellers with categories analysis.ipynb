```python
# Weighted rating
m = min(data['Reviews'])
C = data['User Rating'].mean()
def weighted_rating(x, m=m, C=C):
    v = x['Reviews']
    R = x['User Rating']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
data['Weighted Rating'] = data.apply(weighted_rating, axis=1)
```

```python
wi=data.groupby(['Name','Author','Genre'], as_index=False)[['User Rating', 'Reviews', 'Weighted Rating']].mean().sort_values(by='Weighted Rating', ascending=False).head(10)
```
